```python
# Rotate images
def rotate(image, angle):
        tinggi, lebar = image.shape
        rad = angle*(np.pi/180) 
        
        xmid = tinggi // 2 
        ymid = lebar // 2 
        xnew = tinggi 
        ynew = lebar 
        
        rotate_image = np.zeros((xnew, ynew)) 
        
        for i in range(tinggi):
            for j in range(lebar):
                x = int((i - xmid) * np.cos(rad) - (j - ymid) * np.sin(rad) + xmid)
                y = int((i - xmid) * np.sin(rad) + (j - ymid) * np.cos(rad) + ymid)
                
                if 0 <= x < xnew and 0 <= y < ynew:
                    rotate_image[x, y] = image[i, j]
        return rotate_image

    
# Image Translation and shifting
def translation(image, geser): 
        tinggi, lebar = image.shape[:2]
        imagenew = np.zeros((tinggi, lebar)).astype(int) 

        for y in range(0, 207): 
            for x in range(200, 820): 
                if not 0 <= image[y, x] <= 20: 
                    imagenew[y + geser, x] = image[y, x]

        for y in range(800, tinggi): 
            for x in range(200, 820): 
                if not 0 <= image[y, x] <= 20: 
                    imagenew[y - geser, x] = image[y, x]

        for y in range(300, 720):
            for x in range(0, 210):
                if not 0 <= image[y, x] <= 20:
                    imagenew[y, x + geser] = image[y, x]

        for y in range(300, 720):
            for x in range(820, lebar):
                if not 0 <= image[y, x] <= 20:
                    imagenew[y, x - geser] = image[y, x]

        return imagenew    

def removePadd(img):
        rows = np.any(img > 10, axis=1)
        cols = np.any(img > 10, axis=0)
        return img[rows][:, cols]

def divide_image_vertical(image):
        height, width = image.shape[:2]
        center_y = height // 2
        top_left = image[:center_y, :width//2]
        top_right = image[:center_y, width//2:]
        bottom_left = image[center_y:, :width//2]
        bottom_right = image[center_y:, width//2:]

        return top_left, top_right, bottom_left, bottom_right    

def translate(img, x):
        return np.roll(img, x, axis=1)    

def rotate(img, degree):
        return np.rot90(img, degree//90)    

def crop_image(img):
        rows = np.any(img > 10, axis=1)
        cols = np.any(img > 10, axis=0)
        return img[rows][:, cols]       

# Crop the Image
def cropedImage(image):
        new_image = image[215:812, 309:718] 
        return new_image    
```

```python
# calculating image histograms
def compute_histogram(image):
    histogram = np.zeros(256) 
    for pixel_value in np.nditer(image): 
        histogram[int(pixel_value)] += 1 
    return histogram

# calculate the cumulative distribution function
def compute_cdf(histogram): 
    cdf = np.zeros(256) 
    cdf[0] = histogram[0] 
    for i in range(1, 256):
        cdf[i] = cdf[i-1] + histogram[i]
    return  cdf / cdf.max()

def equalisasi(image):
    histogram = compute_histogram(image)
    cdf = compute_cdf(histogram)
    
    # Normalisasi CDF agar berada dalam rentang [0, 255]
    cdf_normalized = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())
    
    # Menggunakan CDF yang sudah dinormalisasi untuk melakukan transformasi equalisasi
    equalisasi_image = cdf_normalized[image.astype(int)]
    
    return equalisasi_image.astype(np.uint8) if image is not None else None

# menyesuaikan histogram dari gambar asli dengan histogram dari gambar target
def histogram_specification(image, target_image):
    target_hist = compute_histogram(image)
    ref_hist = compute_histogram(target_image)

    target_cdf = compute_cdf(target_hist)
    ref_cdf = compute_cdf(ref_hist)

    mapping = np.zeros(256)
    for i in range(256):
        idx = np.abs(ref_cdf - target_cdf[i]).argmin()
        mapping[i] = idx

    specified_image = mapping[image.astype(np.uint8)]
    
    return specified_image.astype(np.uint8)

#  menyesuaikan histogram dari gambar asli dengan histogram dari gambar target
def match_histograms(source_image, reference_image):
    source_hist, source_bins = np.histogram(source_image.flatten(), bins=256, range=(0, 256), density=True) 
    source_cdf = source_hist.cumsum()
    source_cdf = (255 * source_cdf / source_cdf[-1]).astype(np.uint8)

    reference_hist, reference_bins = np.histogram(reference_image.flatten(), bins=256, range=(0, 256), density=True)
    reference_cdf = reference_hist.cumsum()
    reference_cdf = (255 * reference_cdf / reference_cdf[-1]).astype(np.uint8)

    pixel_mapping = np.zeros(256, dtype=np.uint8)
```

### Pecahan 1x2 

```python
# Fungsi untuk menghitung nilai rata-rata piksel dari suatu segmen citra
def nilai_rata_rata_piksel(image):
    return np.mean(image)

# Fungsi untuk melakukan spesifikasi histogram
def spesifikasi_histogram(image_sumber, image_target):
    # Hitung histogram dari gambar sumber dan target
    hist_sumber, _ = np.histogram(image_sumber.flatten(), bins=256, range=[0,256])
    hist_target, _ = np.histogram(image_target.flatten(), bins=256, range=[0,256])
    
    # Hitung distribusi kumulatif dari histogram gambar sumber dan target
    cdf_sumber = hist_sumber.cumsum()
    cdf_target = hist_target.cumsum()
    
    # Normalisasi distribusi kumulatif agar memiliki nilai maksimum 1
    cdf_sumber = (cdf_sumber - cdf_sumber.min()) / (cdf_sumber.max() - cdf_sumber.min()) * 255
    cdf_target = (cdf_target - cdf_target.min()) / (cdf_target.max() - cdf_target.min()) * 255
    
    # Buat lookup table untuk pemetaan intensitas piksel
    lut = np.interp(cdf_sumber, cdf_target, range(256)).astype(np.uint8)
    
    # Terapkan pemetaan intensitas piksel ke gambar sumber
    matched_image = lut[image_sumber]
    
    return matched_image

# Muat gambar target Kurumi dan Megumin
kurumi_target = cv.imread('kurumi.jpg', cv.IMREAD_GRAYSCALE)
megumin_target = cv.imread('megumin.jpg', cv.IMREAD_GRAYSCALE)

# Muat citra yang akan digabung
combined_image_senhe = cv.imread('combined_image_senhe.jpg', cv.IMREAD_GRAYSCALE)

# Bagi citra menjadi dua bagian horizontal
bagian_atas = combined_image_senhe[:combined_image_senhe.shape[0]//2, :]
bagian_bawah = combined_image_senhe[combined_image_senhe.shape[0]//2:, :]

# Inisialisasi list untuk menyimpan segmen citra yang sudah di-spesifikasi histogramnya
segmen_tertentu = []

# Hitung nilai rata-rata piksel untuk setiap bagian
nilai_rata_rata_atas = nilai_rata_rata_piksel(bagian_atas)
nilai_rata_rata_bawah = nilai_rata_rata_piksel(bagian_bawah)

# Tentukan citra target untuk setiap bagian
citra_target_atas = kurumi_target if nilai_rata_rata_atas <= 128 else megumin_target
citra_target_bawah = kurumi_target if nilai_rata_rata_bawah <= 128 else megumin_target

# Lakukan spesifikasi histogram untuk setiap bagian
hasil_spesifikasi_atas = spesifikasi_histogram(bagian_atas, citra_target_atas)
hasil_spesifikasi_bawah = spesifikasi_histogram(bagian_bawah, citra_target_bawah)

# Gabungkan kembali kedua bagian
hasil_gabungan_citra = np.vstack((hasil_spesifikasi_atas, hasil_spesifikasi_bawah))

# Display original and processed images
plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.imshow(combined_image_senhe, cmap='gray')
plt.title('Original Shenhe Image')
plt.axis('off')

plt.subplot(222)
plt.imshow(hasil_gabungan_citra, cmap='gray')
plt.title('Processed Shenhe Image 1x2')
plt.axis('off')

plt.show()
```

### Pecahan 2x2

```python
# Fungsi untuk menghitung nilai rata-rata piksel dari suatu segmen citra
def nilai_rata_rata_piksel(image):
    return np.mean(image)

# Fungsi untuk melakukan spesifikasi histogram
def spesifikasi_histogram(image_sumber, image_target):
    # Hitung histogram dari gambar sumber dan target
    hist_sumber, _ = np.histogram(image_sumber.flatten(), bins=256, range=[0,256])
    hist_target, _ = np.histogram(image_target.flatten(), bins=256, range=[0,256])
    
    # Hitung distribusi kumulatif dari histogram gambar sumber dan target
    cdf_sumber = hist_sumber.cumsum()
    cdf_target = hist_target.cumsum()
    
    # Normalisasi distribusi kumulatif agar memiliki nilai maksimum 1
    cdf_sumber = (cdf_sumber - cdf_sumber.min()) / (cdf_sumber.max() - cdf_sumber.min()) * 255
    cdf_target = (cdf_target - cdf_target.min()) / (cdf_target.max() - cdf_target.min()) * 255
    
    # Buat lookup table untuk pemetaan intensitas piksel
    lut = np.interp(cdf_sumber, cdf_target, range(256)).astype(np.uint8)
    
    # Terapkan pemetaan intensitas piksel ke gambar sumber
    image_hasil = lut[image_sumber]
    
    return image_hasil

# Muat gambar target Kurumi dan Megumin
kurumi_target = cv.imread('kurumi.jpg', cv.IMREAD_GRAYSCALE)
megumin_target = cv.imread('megumin.jpg', cv.IMREAD_GRAYSCALE)

# Muat citra yang akan digabung
combined_image_senhe = cv.imread('combined_image_senhe.jpg', cv.IMREAD_GRAYSCALE)

# Inisialisasi list untuk menyimpan segmen citra yang sudah di-spesifikasi histogramnya
segmen_tertentu = []

# Bagi citra menjadi empat bagian (2x2)
tinggi, lebar = combined_image_senhe.shape
for i in range(0, tinggi, tinggi // 2):
    for j in range(0, lebar, lebar // 2):
        segmen_citra = combined_image_senhe[i:i + tinggi // 2, j:j + lebar // 2]
        
        # Hitung nilai rata-rata piksel dari segmen citra
        nilai_rata_rata = nilai_rata_rata_piksel(segmen_citra)
        
        # Tentukan citra target berdasarkan nilai rata-rata piksel
        citra_target = kurumi_target if nilai_rata_rata <= 128 else megumin_target
        
        # Lakukan spesifikasi histogram untuk segmen citra
        segmen_tertentu.append(spesifikasi_histogram(segmen_citra, citra_target))

# Gabungkan kembali segmen citra yang sudah di-spesifikasi histogramnya menjadi satu citra
hasil_gabungan_citra = np.vstack([np.hstack(segmen_tertentu[:2]), np.hstack(segmen_tertentu[2:])])

# Display original and processed images
plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.imshow(combined_image_senhe, cmap='gray')
plt.title('Original Shenhe Image')
plt.axis('off')

plt.subplot(222)
plt.imshow(hasil_gabungan_citra, cmap='gray')
plt.title('Processed Shenhe Image 2x2')
plt.axis('off')

plt.show()
```

```python
# Load the Kurumi and Megumin target images
kurumi_target = cv.imread('kurumi.jpg', cv.IMREAD_GRAYSCALE)
megumin_target = cv.imread('megumin.jpg', cv.IMREAD_GRAYSCALE)

# Function to calculate the average pixel value of an image segment
def average_pixel_value(image):
    return np.mean(image)

# Function to perform histogram specification
def histogram_matching(source_image, target_image):
    # Calculate the histograms of the source and target images
    hist_source, _ = np.histogram(source_image.flatten(), bins=256, range=[0,256])
    hist_target, _ = np.histogram(target_image.flatten(), bins=256, range=[0,256])
    
    # Calculate the cumulative distributions of the histograms
    cdf_source = hist_source.cumsum()
    cdf_target = hist_target.cumsum()
    
    # Normalize the cumulative distributions to have a maximum value of 1
    cdf_source = (cdf_source - cdf_source.min()) / (cdf_source.max() - cdf_source.min()) * 255
    cdf_target = (cdf_target - cdf_target.min()) / (cdf_target.max() - cdf_target.min()) * 255
    
    # Create a lookup table for pixel intensity mapping
    lut = np.interp(cdf_source, cdf_target, range(256)).astype(np.uint8)
    
    # Apply the pixel intensity mapping to the source image
    matched_image = lut[source_image]
    
    return matched_image

# Load the combined image
combined_image_senhe = cv.imread('combined_image_senhe.jpg', cv.IMREAD_GRAYSCALE)

# Divide the image into 2 horizontal parts
horizontal_parts = np.array_split(combined_image_senhe, 2, axis=0)

# Initialize a list to store the specified image segments
specified_segments = []

# Iterate over the horizontal parts
for horizontal_part in horizontal_parts:
    # Divide each horizontal part into 16 vertical parts
    vertical_parts = np.array_split(horizontal_part, 16, axis=1)
    
    # Initialize a list to store the specified vertical segments
    specified_verticals = []
    
    # Iterate over the vertical parts
    for vertical_part in vertical_parts:
        # Calculate the average pixel value of the segment
        avg_pixel_value = average_pixel_value(vertical_part)
        
        # Determine the target image based on the average pixel value
        target_image = kurumi_target if avg_pixel_value <= 128 else megumin_target
        
        # Perform histogram specification
        specified_vertical = histogram_matching(vertical_part, target_image)
        
        # Append the specified vertical segment to the list
        specified_verticals.append(specified_vertical)
    
    # Combine the specified vertical segments into a single image
    combined_specified_vertical = np.hstack(specified_verticals)
    
    # Append the combined specified vertical segment to the list
    specified_segments.append(combined_specified_vertical)

# Combine the specified image segments into a single image
combined_specified_image = np.vstack(specified_segments)

# Display original and processed images
plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.imshow(combined_image_senhe, cmap='gray')
plt.title('Original Shenhe Image')
plt.axis('off')

plt.subplot(222)
plt.imshow(combined_specified_image, cmap='gray')
plt.title('Processed Shenhe Image 2x16')
plt.axis('off')

plt.show()
```
